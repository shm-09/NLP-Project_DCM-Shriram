```python
# Extract audio data from the URI link of the uploaded WAV files...
speech_file_uri = "gs://speechtotextfiles_shm09/transcribed_speech_Demo.wav"
```

```python
#Media file loaded from cloud onto variable...
long_audio = s.RecognitionAudio(uri = speech_file_uri)
```

```python
phrase_set = ["fenesta", "dcm shriram", "superbrand", "green building council", "agri rural business", "chloro vinyl", "urea", "sugar", "farm solutions", "hariyali kisaan bazaar", "hybrid seeds", "alkali", "calcium carbide", "PVC", "ultra luxury", "super luxury", "luxury", "slimline slider", "lift and slide", "slide and fold", "slider", "casement", "triple sealing", "multi-point"]
```

```python
# Configure media file output...MP3 config used...
config_wav = {"language_code": "en-US", 
"enable_automatic_punctuation": True,
"model": "video",
"use_enhanced": True,
"audio_channel_count": 2,
"speech_contexts": [{"phrases": phrase_set, "boost": 10.0}]}

# Note: We can use enchance =True & Model = Video arguments in the Config function for better/more accurate transcription if required... 
```

```python
response_wav = client.long_running_recognize(config = config_wav, audio = long_audio)
```

```python
temp = response_wav.result()
# Note: takes a while to execute... (around 6 mins per video)
```

```python
for result in temp.results:
    f = open("Transcripts\\transcript_Suresh.txt", "a")
    f.write(result.alternatives[0].transcript)
    f.write(" ")
f.close()
```

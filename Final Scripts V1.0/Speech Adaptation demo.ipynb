```python
transcribed_audio_file = "Audio_extracted\\transcribed_speech_Ayush Kumar.wav"
video_file = "Videos\\Ayush - Ayush Sharma (West Central Sales).mp4"
audioclip = AudioFileClip(video_file)
audioclip.write_audiofile(transcribed_audio_file)

# Upload the list of transcribed audio files onto google cloud for further steps...
```

```python
# Extract audio data from the URI link of the uploaded WAV files...
speech_file_uri = "gs://speechtotextfiles_shm09/transcribed_speech_Demo.wav"
```

```python
# Define the phrase sets to be used here for transcript boosting...

phrase_set = ["fenesta", "dcm shriram", "superbrand", "green building council", "agri rural business", "chloro vinyl", "urea", "sugar", "farm solutions", "hariyali kisaan bazaar", "hybrid seeds", "alkali", "calcium carbide", "PVC", "ultra luxury", "super luxury", "luxury", "slimline slider", "lift and slide", "slide and fold", "slider", "casement", "triple sealing", "multi-point"]
```

```python
# putting in the config parameters of the transcription required...

config = {"language_code": "en-US", 
"enable_automatic_punctuation": True,
"model": "video",
"use_enhanced": True,
"audio_channel_count": 2,
"speech_contexts": [{"phrases": phrase_set, "boost": 10.0}]}
```

```python
long_audio = s.RecognitionAudio(uri = speech_file_uri)
```

```python
response = client.long_running_recognize(config = config, audio = long_audio)
```

```python
temp = response.result()
```

```python
# Saving the transcripts in a text file...

for result in temp.results:
    f = open("Transcripts\\transcript_Ayush Kumar.txt", "a")
    f.write(result.alternatives[0].transcript)
    f.write(" ")
f.close()
```
